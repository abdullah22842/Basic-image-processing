```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from skimage.measure import label, regionprops_table
import statistics
from scipy import ndimage
```

```python
img_path = 'Rectangle and Circle Image.BMP'
img = cv2.imread(img_path)
img.shape
```

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
bw = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)[1]
```

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(bw, cmap='gray')
plt.show()
```

```python
label_img =label(bw)
```

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(label_img, cmap='gray')
plt.show()
```

```python
props = regionprops_table(label_img, properties=('area',
                                                 'perimeter',
                                                 'major_axis_length',
                                                 'minor_axis_length' ))
```

```python
measure = pd.DataFrame(props)
measure
```

```python
Avg_Area = statistics.mean(measure['area'])
Average_Perimeter = statistics.mean(measure['perimeter'])
Average_Major_Axis_Length = statistics.mean(measure['major_axis_length'])
Average_Minor_Axis_Length = statistics.mean(measure['minor_axis_length'])
print(Avg_Area)
print(Average_Perimeter)
print(Average_Major_Axis_Length)
print(Average_Minor_Axis_Length)
```
