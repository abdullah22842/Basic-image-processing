## filter

```python
arr = np.empty(350 * 350)

for idx in range(350 * 350):
  if idx % 2 == 0 :
    arr[idx] = 0
  else:
    arr[idx] = 1
```

```python
arr = arr.reshape((350, 350))
arr
```

```python
plt.imshow(arr, vmin=0, vmax=1, cmap='gray')
plt.grid(None)
plt.show()
```

```python
plt.imshow(dog_gray * arr, vmin=0, vmax=255, cmap='gray')
plt.grid(None)
plt.show()
```
